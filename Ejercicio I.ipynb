**6. Es cierto que la gente que gana mas que 50K tiene al menos educacion media completa? (variable *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate*)**

```python
array = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
data.loc[(data['salary'] == '>50K') & data['education'].isin(array)]
```

Esto está mal, porque precisamente estoy viendo las personas que ganas mas de 50K al año y tienen al menos educación media completa. Podria contarlos y compararlos con los que ganan mas de 50K al año sin filtro alguno pero esto suena muy ineficiente. 
Si lo puedo usar para comprobar el resultado, que como se puede ver, me debería dar que hay $(7841-4535)=3306$ personas que ganan mas de 50K al año sin tener al menos el nivel medio de educación completo.

```python
array = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
data.loc[(data['salary'] == '>50K') & ~data['education'].isin(array)]
```
